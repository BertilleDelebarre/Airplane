```python
flight_indent_list = ["CMP435","CMP392","CMP236","CMP304","CMP312","CMP809"]
#flight_indent_list = ["CMP435"]
z_total = pd.DataFrame()

for flight_indent in flight_indent_list:
    z1 = first_line[first_line.ident == flight_indent][first_line.type =="position"]
    z1 = z1[["lat","lon","clock"]]
    print(z1.shape)
    z1["fill"] = "red"
    z2 = first_line[first_line.ident == flight_indent][first_line.type =="flightplan"][first_line.facility_name =="Airline"]
    z2 = pd.DataFrame(z2.waypoints.values[0])
    z2["fill"] = "blue"
    z2["clock"] = "NaN"
    z_total = pd.concat((z_total,z1))
    z_total = pd.concat((z_total,z2))
```
