```python
#flight_ident_list = ["CMP435","CMP392","CMP236","CMP304","CMP312","CMP809"]
FeatureCollection = {"type" : "FeatureCollection",
           "features" : []}

for flight_ident in flight_ident_list :
    unique_id = z_total[z_total.ident == flight_ident ]["id"].unique()
    for ids in unique_id :
        Feature = { "type": "Feature",
            "geometry": {
            "type": "LineString",
            "coordinates": [] 
            },
            "properties": {}
         }
        Feature["properties"]["id"] = ids
        Feature["properties"]["ident"] = flight_ident
        Feature["properties"]["fill"] = z_total[z_total.id == ids]["fill"].iloc[0]
        Feature["properties"]["clock"] = z_total[z_total.id == ids]["clock"].values.tolist()
        Feature["geometry"]["coordinates"] = z_total[z_total.id == ids][["lon","lat"]].values.tolist()
        FeatureCollection["features"] += [Feature]
```

```python
with open("Line_map/data_file2.json", "w") as write_file:
    json.dump(FeatureCollection, write_file)
```
